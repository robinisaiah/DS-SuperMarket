```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Product line', data=data)
plt.xticks(rotation=45)
plt.title("Sales Count by Product Line")
plt.show()
```

```python
city_sales = data.groupby('City')['Total'].sum()
city_sales.plot(kind='bar', color='skyblue')
plt.title("Total Sales by City")
plt.ylabel("Sales in USD")
plt.show()
```

```python
sns.countplot(x='Payment', data=data, palette="viridis")
plt.title("Payment Method Distribution")
plt.show()
```

```python
avg_rating = data.groupby('Product line')['Rating'].mean()
avg_rating.plot(kind='bar', color='lightgreen')
plt.title("Average Customer Rating by Product Line")
plt.ylabel("Average Rating")
plt.show()

```

```python
sns.scatterplot(x='Quantity', y='Total', data=data, alpha=0.5)
plt.title("Quantity vs Total Sales")
plt.show()

```
